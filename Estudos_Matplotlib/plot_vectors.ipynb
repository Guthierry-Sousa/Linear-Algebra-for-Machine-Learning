# Plot Vectors

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
vector = np.array([2,3])
origem = np.array([0,0])

plt.quiver(origem[0], origem[1], vector[0], vector[1], angles='xy', scale_units='xy', scale=1)

label = f"({vector[0]}, {vector[1]})"
plt.text(vector[0] + 0.1, vector[1] + 0.1, label, fontsize = 13)

plt.xlim(-1, 5)
plt.ylim(-1, 5)
plt.axhline(0)
plt.axvline(0)
plt.grid()

plt.show()
```

```python
x = np.array([0, 0, 0])
y = np.array([0, 0, 0])

vx = np.array([2, 1, 3])
vy = np.array([1, 2, 1])

plt.quiver(x, y, vx, vy, angles='xy', scale_units='xy', scale=1)

plt.xlim(-1, 4)
plt.ylim(-1, 4)
plt.axhline(0,color = 'red')
plt.axvline(0,color = 'red')
plt.grid()
plt.show()
```

```python
# Soma de vetores
o = np.array([0,0])

a = np.array([2,3])
b = np.array([-1,2])

c = a + b

#Multiplicação por escalar

d = -1*a
```

```python
# Vetor a
label_a = f'({a[0]}, {a[1]})'
plt.quiver(o[0], o[1], a[0], a[1], angles='xy', scale_units='xy', scale=1, color = 'red')
plt.text(a[0]+0.1, a[1]+0.1, label_a, fontsize = 12)

#Vetor b:
label_b = f'({b[0]}, {b[1]})'
plt.quiver(o[0], o[1], b[0], b[1], angles='xy', scale_units='xy', scale=1, color = 'blue')
plt.text(b[0]+0.1, b[1]+0.1, label_b, fontsize = 12)

#Vetor c (a+b):
label_c = f'({c[0]}, {c[1]})'
plt.quiver(o[0], o[1], c[0], c[1], angles='xy', scale_units='xy', scale=1, color = 'orange')
plt.text(c[0]+0.1, c[1]+0.1, label_c, fontsize = 12)

# Vetor d (-1a)
label_d = f'({d[0]}, {d[1]})'
plt.quiver(o[0], o[1], d[0], d[1], angles='xy', scale_units='xy', scale=1, color = 'green')
plt.text(d[0]+0.3, d[1]+0.1, label_d, fontsize = 12)

plt.xlim(-4,6)
plt.ylim(-4,6)
plt.axhline(0, color = 'black')
plt.axvline(0, color = 'black')
plt.grid()
plt.show()
```
