# Plot Vectors

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
vector = np.array([2,3])


origem = np.array([0,0])

plt.quiver(origem[0], origem[1], vector[0], vector[1], angles='xy', scale_units='xy', scale=1)

label = f"({vector[0]}, {vector[1]})"
plt.text(vector[0] + 0.1, vector[1] + 0.1, label, fontsize = 13)

plt.xlim(-1, 5)
plt.ylim(-1, 5)
plt.axhline(0)
plt.axvline(0)
plt.grid()

plt.show()
```

```python
x = np.array([0, 0, 0])
y = np.array([0, 0, 0])

vx = np.array([2, 1, 3])
vy = np.array([1, 2, 1])

plt.quiver(x, y, vx, vy, angles='xy', scale_units='xy', scale=1)

plt.xlim(-1, 4)
plt.ylim(-1, 4)
plt.axhline(0,color = 'red')
plt.axvline(0,color = 'red')
plt.grid()
plt.show()
```
