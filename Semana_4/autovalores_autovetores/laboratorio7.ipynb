### **Interpreting eigenvalues and eigenvectors**

```python
import numpy as np
import matplotlib.pyplot as plt
```

Let's consider a linear transformation defined by matrix $A=\begin{bmatrix}2 & 3 \\ 2 & 1 \end{bmatrix}$. Apply this transformation to the standard basis vectors $e_1=\begin{bmatrix}1 \\ 0\end{bmatrix}$ and $e_2=\begin{bmatrix}0 \\ 1\end{bmatrix}$ and visualize the result. Hopefully using a matrix to transform a basis of vectors is familiar from earlier lectures in the course.

```python
# Transformação Linear:

def T(x):

    A = np.array([[2,3],[2,1]])

    w = A @ x

    return w

# Aplicando nos vetores da base

e_1 = np.array([1,0])
e_2 = np.array([0,1])

# Visualizando o resultado

print(f"T(e_1) = {T(e_1)}")
print(f"T(e_2) = {T(e_2)}")
```

```python
# Plotando transformações lineares

def plot_transformation(T, e1, e2, label_vector = 'v'):

    plt.figure(figsize=(10,8))
    plt.xlim(-5,5)
    plt.ylim(-5,5)

    plt.quiver(0, 0, e1[0], e1[1], angles='xy', scale_units='xy', scale=1, color = 'red')
    plt.text(e1[0] + 0.1, e1[1] + 0.1, f"${label_vector}_1$", fontsize = 13)

    plt.quiver(0, 0, e2[0], e2[1], angles='xy', scale_units='xy', scale=1, color = 'red')
    plt.text(e2[0] + 0.1, e2[1] + 0.1, f"${label_vector}_2$", fontsize = 13)

    T_e1 = T(e1)
    T_e2 = T(e2)

    plt.quiver(0, 0, T_e1[0], T_e1[1], angles='xy', scale_units='xy', scale=1, color = 'blue')
    plt.text(T_e1[0] + 0.1, T_e1[1] + 0.1, f"$T({label_vector}_1)$", fontsize = 13)

    plt.quiver(0, 0, T_e2[0], T_e2[1], angles='xy', scale_units='xy', scale=1, color = 'blue')
    plt.text(T_e2[0] + 0.1, T_e2[1] + 0.1, f"$T({label_vector}_2)$", fontsize = 13)

    plt.grid()
    plt.show()
```

```python
plot_transformation(T, e_1, e_2, 'e')
```

```python
# Encontrando autovalores e autovetores
A = np.array([[2,3],[2,1]])

A_eigenvalues, matriz_eigenvectors = np.linalg.eig(A)

print(f"Matriz A: \n{A}\n")
print(f"Autovalores de A: {A_eigenvalues}\n")
print(f"Matriz de autovetores: \n{matriz_eigenvectors}")
```

```python
plot_transformation(T, matriz_eigenvectors[:,0], matriz_eigenvectors[:,1])
```

### **Example 1: Reflection about y-axis (the vertical axis)**

```python
# Matriz de transformação linear:
# T(1,0) = (-1,0)
# T(0,1) = (0,1) 

def T_reflection(v):

    M = np.array([[-1, 0],[0, 1]])

    w = M @ v

    return w

e1 = np.array([1,0])
e2 = np.array([0,1])

plot_transformation(T_reflection, e1, e2, 'e')
```

```python
M = np.array([[-1, 0],[0, 1]])

M_eigenvalues, M_eigenvectors = np.linalg.eig(M)

print(f"Matriz M: \n{M}\n")
print(f"Autovalores de M: \n{M_eigenvalues}\n")
print(f"Matriz de autovetores: \n{M_eigenvectors}")

plot_transformation(T_reflection, M_eigenvectors[:,0], M_eigenvectors[:,1])
```

### **Example 2: Shear in x-direction**

```python
def T_shear(v):

    S = np.array([[1, 0.5],[0,1]])

    w = S @ v

    return w

plot_transformation(T_shear, e1, e2,  'e')
```

```python
S_shear = np.array([[1,0.5],[0,1]])

S_eigenvalues, S_eigenvectors = np.linalg.eig(S_shear)

print(f"Matriz S: \n{S_shear}\n")
print(f"Autovalores de S: \n{S_eigenvalues}\n")
print(f"Matriz de autovetores: \n{S_eigenvectors}")

plot_transformation(T_shear, S_eigenvectors[:,0], S_eigenvectors[:,1])
```

```python
# [cos -sen]
# [sen cos]

def T_rotation_90(v):

    R = np.array([[0, -1],[1,0]]) # Sem autovalores reais

    w = R @ v

    return w

def T_rotation_180(v):

    R = np.array([[-1, 0],[0, -1]]) # Possui autovalores reais

    w = R @ v

    return w


plot_transformation(T_rotation_90, e1, e2, 'e')
plot_transformation(T_rotation_180, e1, e2, 'e')
```

### **Example 3: Identity Matrix and Scaling in All Directions**

What happens if we transform the plane using the identity matrix? This means that there will be no change to any vector in the plane. Since every point and vector does not move at all, in this case every vector is still facing in the same direction, and every vectors meets the definition of an eigenvector.

```python
A_identity = np.array([[1,0],[0,1]])
A_eigenvalues, A_eigenvectors = np.linalg.eig(A_identity)

print(f"Matriz A: \n{A_identity}\n")
print(f"Autovalores de A: \n{A_eigenvalues}\n")
print(f"Matriz de autovetores: \n{A_eigenvectors}")

# Os autovalores são os valores da própria diagonal principal
```

```python
def T_projection_x(v):

    A_projection = np.array([[1,0],[0,0]])

    w = A_projection @ v

    return w

plot_transformation(T_projection_x, e1, e2, 'e')
```

```python
A_projection = np.array([[1,0],[0,0]])
A_eigenvalues, A_eigenvectors = np.linalg.eig(A_projection)

print(f"Matriz A: \n{A_projection}\n")
print(f"Autovalores de A: \n{A_eigenvalues}\n")
print(f"Matriz de autovetores: \n{A_eigenvectors}")

plot_transformation(T_projection_x, A_eigenvectors[:,0], A_eigenvectors[:,1])
```
