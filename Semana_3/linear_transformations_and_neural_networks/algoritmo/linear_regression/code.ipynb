# Linear Regression



```python
from sklearn.linear_model import LinearRegression
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
dados = {
    "horas_estudadas": [1, 2, 2.5, 3, 3.5, 4, 4.5, 5, 5.5, 6,
                        6.5, 7, 7.5, 8, 8.5, 9, 9.5, 10],
    "nota_obtida":     [45, 50, 52, 55, 58, 60, 62, 65, 67, 70,
                        72, 75, 78, 80, 82, 85, 88, 90]
}

df_dados = pd.DataFrame(dados)
display(df_dados)
```

```python
# Plotando valores

plt.title("Horas Estudadas x Nota Obtida")
plt.scatter(df_dados['horas_estudadas'], df_dados['nota_obtida'], color = 'black')
plt.xlabel('Horas Estudadas')
plt.ylabel('Notas Obtidas')

plt.grid()
plt.show()
```

```python
X = np.array(df_dados['horas_estudadas']).reshape(-1,1)
print(f"X: \n{X}")

y = np.array(df_dados['nota_obtida'])
print(f"\ny = {y}")
```

```python
model = LinearRegression().fit(X, y)

y_pred = model.predict(X)
b0 = model.intercept_
b1 = model.coef_

# Valores Previstos
print(f"Predict: {y_pred}\n")

# Intercepto y
print(f"b0: {b0}\n")

# Inclinação (slope)
print(f"b1: {b1}")

df_dados['y_predict'] = y_pred

display(df_dados)
```

```python
plt.title("Horas Estudadas x Nota Obtida")
plt.scatter(df_dados['horas_estudadas'], df_dados['nota_obtida'], color = 'black')
plt.plot(X, y_pred, color = 'red')
plt.xlabel('Horas Estudadas')
plt.ylabel('Notas Obtidas')

plt.grid()
plt.show()
```

```python
def function_reg(x):

    f = b0 + b1 * x

    return f

dados_teste = {
    "horas_estudadas": [1.5, 3.2, 4.8, 6.3, 7.8, 9.2],
    "nota_obtida":     [48, 56, 63, 71, 79, 87]
}

df_teste = pd.DataFrame(dados_teste)

display(df_teste)

X_teste = np.array(df_teste['horas_estudadas']).reshape(-1,1)
y_pred_teste = function_reg(X_teste)

df_teste['y_pred'] = y_pred_teste

display(df_teste)
```

#### Regressão Linear de forma prática

```python
from sklearn.datasets import load_diabetes

data = load_diabetes()
print(data)
```

```python
df = pd.DataFrame(data.data)
df.columns = data.feature_names
df['progressao_diabetes'] = data.target

display(df)
```

```python
from sklearn.model_selection import train_test_split

X = np.array(df.age).reshape(-1,1)
y = np.array(df.progressao_diabetes)

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.30, random_state=42)
```

```python
# Treino

model = LinearRegression().fit(X_train, y_train)

b0 = model.intercept_
print(f"b0: {b0}")

b1 = model.coef_
print(f"b1: {b1}")

predict_t = model.predict(X_test)
print(f"y_pred: {predict_t}")
```

```python

plt.scatter(X_train, y_train)
plt.scatter(X_test, y_test)

plt.plot(X_test, predict_t, color = 'red')
```

```python
from sklearn.metrics import mean_squared_error, r2_score

mse = mean_squared_error(y_test, predict_t)
r2 = r2_score(y_test, predict_t)

print(f"MSE: {mse}")
print(f"R²: {r2}")
```
