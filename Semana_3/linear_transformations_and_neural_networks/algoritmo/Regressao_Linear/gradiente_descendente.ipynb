Algoritmo - Gradiente Descendente

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def gradient_descent(x, y, n):

    b = 0
    w = 0

    learning_rate = 0.01

    for _ in range(n):

        f = w * x + b

        w = w - (learning_rate * np.mean((f - y) * x))

        b = b - (learning_rate * np.mean(f-y))

    return w, b
```

```python
X = np.array([2,4,5])
y = np.array([10,20,40])

w , b = gradient_descent(X, y, 10000)

print(f'coef: {w}')
print(f'intercept: {b}')
```

```python
plt.plot(X, w*X+b)
plt.scatter(X, y)

plt.show()
```
