## Vetores e Matrizes

```python
import numpy as np
```

### Operações vetoriais

##### 1- Soma e Subtração

```python
# Os vetores precisam ter a mesma dimensão. A soma acontece elemento por elemento.

vetor_1 = np.array([1,2,3])
vetor_2 = np.array([4,3,5])

print(f"Vetor 1: {vetor_1}")
print(f"\nVetor 2: {vetor_2}")

vetor_soma = vetor_1 + vetor_2
print(f"\n{vetor_1} + {vetor_2} = {vetor_soma}")
```

```python
vetor_subtracao = vetor_1 - vetor_2
print(f"\n{vetor_1} + {vetor_2} = {vetor_subtracao}")
```

##### 2- Multiplicação por escalar

```python
# Nessa operação cada componente do vetor é multiplicado por um escalar (número) qualquer.

escalar1 = 3
escalar2 = -0.5

v = np.array([3,4,5])

print(f"{escalar1} x {v} = {escalar1 * v}")
print(f"\n{escalar2} x {v} = {escalar2 * v}")
```

### Norma de um vetor

```python
# A norma é basicamente o módulo (tamanho) do vetor, ou seja, a distância da origem até sua coordenada.
# Definição: Ela representa a distância euclidiana da origem do sistema de coordenadas (o ponto zero) até a coordenada final onde o vetor termina.

u = np.arange(1,10,0.5)
print("Vetor u:")
print(u)

norma_u = np.linalg.norm(u)

print(f'Norma de u: {norma_u:.2f}')
```

```python
#Exemplo: Calcule a distância entre os vetores a = (3,6) e b = (5,2)

a = np.array([3,6])
b = np.array([5,2])

vetor_subtracao = a-b
print(f"{a} - {b} = {vetor_subtracao}")

#Distância: ||a-b||

d = np.linalg.norm(vetor_subtracao)

print(f"Distância entre {a} e {b} = {d:.2f}")
```

### Produto Escalar/Interno

```python
x = np.random.randint(1,5,5)
print(x)
y = np.random.randint(1,5,5)
print(y)

# Produto Escalar = <x,y> = x1*y1 + ... +xn*yn

dot_prod = np.dot(x,y)
print(f"<x,y> = {dot_prod}")
```

### Ângulo entre vetores

```python
u = np.array([1,2,3,4])

v = np.array([4,3,2,5])

dot_prod = np.dot(u,v)

norm_u = np.linalg.norm(u)
norm_v = np.linalg.norm(v)

print(f"Dot = {dot_prod}")
print(f"Norma u = {norm_u}")
print(f"Norm v = {norm_v}")
```

```python
angulo_vetores_rad = np.arccos(dot_prod/(norm_u*norm_v))

angulo_vetores_graus = (angulo_vetores_rad*180)/np.pi

print(f"Angulo entre os vetores {u} e {v} = Rad:{angulo_vetores_rad} | Graus:{angulo_vetores_graus}")
```
